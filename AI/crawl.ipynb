```python
import requests
from bs4 import BeautifulSoup

url = "https://dulichviet.com.vn/du-lich-trong-nuoc"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    tour_boxes = soup.find_all("div", class_="mda-box-item")
    
    for tour_box in tour_boxes:
        tour_name = tour_box.find("h3").text.strip()
        tour_description = tour_box.find("div", class_="des").text.strip()
        tour_time = tour_box.find("p", class_="mda-time-grid").text.strip()
        tour_price = tour_box.find("span", class_="mda-money-red").text.strip()

        # Extract departure location and hotel category with error handling
        tour_caption = tour_box.find("div", class_="mda-caption")
        departure_location_elem = tour_caption.find("div", class_="mda-box-lb")
        departure_location = departure_location_elem.text.strip() if departure_location_elem else "Không có thông tin điểm khởi hành"
        
        hotel_category_elem = tour_caption.find("span", class_="mda-day")
        hotel_category = hotel_category_elem.text.strip() if hotel_category_elem else "Không có thông tin loại khách sạn"

        # Tách thông tin về thời gian, ngày khởi hành và vận chuyển
        tour_info = tour_box.find("div", class_="mda-info mda-grid")
        tour_schedule_elem = tour_info.find("p", class_="mda-schedule")
        tour_schedule = tour_schedule_elem.text.strip() if tour_schedule_elem else "Không có thông tin lịch trình"
        
        tour_transport_elem = tour_info.find("p", class_="mda-time-grid")
        tour_transport = tour_transport_elem.text.strip() if tour_transport_elem else "Không có thông tin vận chuyển"
        
        # Extract image URL
        image_url = tour_box.find("img")["src"] if tour_box.find("img") else "Không có ảnh"

        print("Tên tour:", tour_name)
        print("Mô tả tour:", tour_description)
        print("Thời gian:", tour_time)
        print("Giá tiền:", tour_price)
        print("Điểm khởi hành:", departure_location)
        print("Loại khách sạn:", hotel_category)
        print("Lịch trình:", tour_schedule)
        print("Vận chuyển:", tour_transport)
        print("URL ảnh:", image_url)
        print("--------------------")
else:
    print("Không thể truy cập trang:", url)
```
